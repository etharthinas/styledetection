```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import re
import nltk.data
import nltk
```

```python
nltk.download('punkt')
sent_tokenizer = nltk.data.load("tokenizers/punkt/english.pickle")
```

```python
data = pd.read_csv("blogtext.csv")
```

```python
data["ga"] = (np.round((data.age-5)/10)*10).astype(int).map(str) + data.gender
df = data[["id", "ga", "topic", "text"]]
df.rename(columns = {"id": "author_id"}, inplace = True)
```

```python
splitted = df["text"].apply(lambda x: sent_tokenizer.tokenize(x)).explode()
```

```python
sentences = pd.DataFrame(splitted)
sentences.reset_index(inplace = True)
sentences.rename(columns = {"index": "post_id"}, inplace = True)
sentences.text = sentences.text.fillna("")
sentences.to_csv("sentences.csv")
```
